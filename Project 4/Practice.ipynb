```python
file="tweet_json.txt"
import json
def read_json(file):
    data=[]
    with open(file) as json_file:
        for line in json_file:
            data.append(line)
    return data
        
```

```python
data=read_json(file)
```

```python
json_obj=json.loads(data[0])
```

```python
json_obj["created_at"]
```

```python
def dataframe_json(file):
    data=read_json(file)
    length=len(data)
    json_data=[{"id":json.loads(data[k])["id"],"retweet_count":json.loads(data[k])["retweet_count"],
                "favorite_count":json.loads(data[k])["favorite_count"],
               "created_at":json.loads(data[k])["created_at"]} for k in range(length)] 
    df_json=pd.DataFrame(json_data)
    return df_json
   
```
